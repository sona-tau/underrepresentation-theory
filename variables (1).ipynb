```python
class Source:
    def __init__(s, title: str, location: str, years: list[int], variables: list[str]):
        s.title = title
        s.location = location
        s.years = years
        s.variables = variables

    def dict(s):
        return {
            "title": s.title,
            "location": s.location,
            "years": s.years,
            "variables": s.variables
        }
```
